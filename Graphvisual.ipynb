```python
import networkx as nx
import matplotlib.pyplot as plt

#defining the class
class GraphVisualization:
  def __init__(self):
    self.visual = []

  def addEdge(self,a,b):
      temp = [a,b]
      self.visual.append(temp)

  def visualize(self):
    G=nx.Graph()
    G.add_edges_from(self.visual)
    nx.draw_networkx(G)
    plt.show()
G = GraphVisualization()
G.addEdge(0 ,2)
G.addEdge(1 ,2)
G.addEdge(1, 3)
G.addEdge(5, 3)
G.addEdge(3, 4)
G.addEdge(1 ,0)
G.visualize()
```
