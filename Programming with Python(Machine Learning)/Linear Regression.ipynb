```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
plt.scatter(x,y)
plt.xlabel("Years after Purchase")
plt.ylabel("Resale Value")
plt.show()
```

```python
slope,intercept,r,p,std_error=stats.linregress(x,y)
def myfunc(x):
    return slope*x+intercept
my_model=list(map(myfunc,x))
```

```python
print(slope)
print(intercept)
print(std_error)
print(r)
print(my_model)
```

```python
plt.scatter(x,y)
plt.xlabel("Years after purchase")
plt.ylabel("Reasale Value")
plt.plot(x,my_model,lw=4,c='Orange')
plt.show()
```
