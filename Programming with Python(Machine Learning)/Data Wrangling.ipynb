```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
!wget =  https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename , names = headers)
```

```python
df.head(5)
```

```python
#Replacing The ? with NaN 
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Checking if there are any null data in the form of True and false
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Using For-loop to check the missing values in all the columns
for column in missing_data.columns.values.tolist():
    print(column)#prints the coloumns
    print(missing_data[column].value_counts()) #values_counts checks how many missing values are missing true = missing
    print("")
```

```python
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis = 0)
print("The Average Normalized loss", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
avg_bore = df['bore'].astype('float').mean(axis = 0)
print("The Average bore ", avg_bore)
```

```python
df['bore'].replace(np.nan, avg_bore, inplace = True)
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("The Average Stroke is ",avg_stroke)
df['stroke'].replace(np.nan, avg_stroke,inplace = True)
```

```python
avg_hp = df['horsepower'].astype(float).mean(axis = 0)
print("The average Horse Power is",avg_hp)
df['horsepower'].replace(np.nan, avg_hp, inplace = True)
```

```python
avg_price = df['price'].astype('float').mean(axis = 0)
print("The Average Price is ", avg_price)
df['price'].replace(np.nan, avg_price, inplace = True)
```

```python
avg_prpm = df['peak-rpm'].astype('float').mean(axis = 0)
print("The average peak-rpm is ",avg_prpm)
df['peak-rpm'].replace(np.nan, avg_prpm, inplace = True)
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, "four",inplace = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df[['bore','stroke']] = df[['bore','stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int64')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
df['highway-L/100km'] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'},inplace = True)
df.head()
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['length']/df['length'].max()
df[['length','width','height']].head()
```

```python
df['horsepower'] = df['horsepower'].astype(int, copy = True)
```

```python
df['horsepower'].hist()
plt.xlabel('horsepower')
plt.ylabel('counts')
plt.title("horsepower bins")
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'],bins, labels = group_names, include_lowest = True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.hist(df['horsepower'], bins = 3)
plt.xlabel("horsepower")
plt.ylabel("counts")
plt.title("Horsepower bins")
```

```python
df.columns
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas','diesel':'fuel-type-diesel'},inplace = True)
dummy_variable_1.head(20)
```

```python
df = pd.concat([df, dummy_variable_1], axis = 1)
df.drop("fuel-type", axis = 1, inplace = True)
df.head()
```

```python
dummy_var = pd.get_dummies(df['aspiration'])
dummy_var
```

```python
dummy_var.rename(columns={'std':'asp-type-std', 'turbo':'asp-type-turbo'},inplace = True)
dummy_var.head(20)
```

```python
df = pd.concat([df, dummy_var], axis = 1)
df.drop(['aspiration'],axis = 1, inplace = True)
df.head()
```

```python
df.to_csv('clean-df.csv')
```
